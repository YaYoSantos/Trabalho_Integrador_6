# Fase 3 — Normalização (Silver)

Lê os arquivos brutos do Bronze, normaliza tipos e salva CSVs padronizados em `silver/`.

**Pré-requisito:** `01_coleta.ipynb` já executado (bronze populado).

```python
import sys, pathlib

_root = pathlib.Path.cwd()
if not (_root / "config.py").exists():
    _root = _root.parent
if str(_root) not in sys.path:
    sys.path.insert(0, str(_root))
```

```python
from config import config

print(f"Bronze dir : {config.bronze_dir}")
print(f"Silver dir : {config.silver_dir}")
```

```python
from silver import PnadNormalizer

print("=== Fase 3.1 — Normalizar PNAD ===")
pnad_norm = PnadNormalizer(config.bronze_dir, config.silver_dir)
df_pnad = pnad_norm.normalize()
df_pnad.head()
```

```python
from silver import CagedNormalizer

print("=== Fase 3.2 — Normalizar CAGED ===")
caged_norm = CagedNormalizer(config.bronze_dir, config.silver_dir)
df_caged = caged_norm.normalize()
df_caged.head()
```

```python
from silver import RaisNormalizer

print("=== Fase 3.3 — Normalizar RAIS ===")
rais_norm = RaisNormalizer(config.bronze_dir, config.silver_dir)
df_rais = rais_norm.normalize()
df_rais.head()
```
